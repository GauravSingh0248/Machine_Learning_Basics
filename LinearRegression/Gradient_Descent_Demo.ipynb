```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate data using dummy function
X=np.arange(10)
Y=(X-5)**2+3
print(X,Y)
```

```python
# Visualize 
plt.style.use("seaborn-v0_8")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Gradient Descent
# intialize x with any random value 
x=9
lr=0.1
plt.plot(X,Y)
for i in range(50):
    grad=2*(x-5)
    x=x-lr*grad
    y=(x-5)**2+3
    plt.scatter(x,y)
    print(x)
plt.show()
```
